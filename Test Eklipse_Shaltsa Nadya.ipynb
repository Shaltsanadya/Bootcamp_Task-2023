```python
import pandas as pd
```

```python
df= pd.read_csv('clips.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
df1= pd.read_csv('downloaded_clips.csv')
pd.set_option('display.max_columns', None)
df1.head()
```

```python
df2= pd.read_csv('gamesession.csv')
pd.set_option('display.max_columns', None)
df2.head()
```

```python
df3= pd.read_csv('premium_users.csv')
pd.set_option('display.max_columns', None)
df3.head()
```

```python
df4= pd.read_csv('shared_clips.csv')
pd.set_option('display.max_columns', None)
df4.head()
```

```python
df3['status']= 'Premium'
group= pd.DataFrame(df3.user_id.groupby(df3['user_id']).count())
premium= pd.DataFrame()
premium['user_id']= group.index
premium['status']= df3['status']
premium
```

```python
df1.rename(columns={'user_Id':'user_id'}, inplace=True)
```

```python

#downloaded_clips file
download_clip= df1.merge(premium, on= 'user_id', how= 'left')
download_clip['status'].fillna('Free', inplace=True)
#gamesession file
gamesession= df2.merge(premium, on= 'user_id', how= 'left')
gamesession['status'].fillna('Free', inplace=True)
#shared_clips file
shared_clips= df4.merge(premium, on= 'user_id', how= 'left')
shared_clips['status'].fillna('Free', inplace=True)

#hapus Unnamed: 0
list= [download_clip, gamesession, shared_clips]
for i in list:
    i.drop(['Unnamed: 0'], axis=1, inplace=True)


```

```python
#save file
download_clip.to_csv('download_clip.csv', index= False)
gamesession.to_csv('sesigame.csv', index= False)
shared_clips.to_csv('share.csv', index= False)
```

```python
user= df.merge(premium, on='user_id', how='left')
user['status'].fillna('Free', inplace=True)
```

```python
user.head()
```

```python
user.drop(['Unnamed: 0'], axis=1, inplace=True)
user['created_at']= pd.to_datetime(user['created_at'])
user['join_at']= pd.to_datetime(user['join_at'])
user['created_hour']= user['created_at'].dt.hour
user['created_day']= user['created_at'].dt.day_name()
user['join_hour']= user['join_at'].dt.hour
user['join_day']= user['join_at'].dt.day_name()
```

```python
user.isnull().sum()/len(user)*100
```

```python
user.dropna(inplace=True)
```

```python
user.columns
```

```python
user.head()
```

```python
replacement= {1: 'horizontal AI highlight clips', 
       2: 'converted to TikTok clips',
       3: 'trimmed clips',
       5: 'eventful highlight clips based on stream statistic',
       6: 'weekly montage clips created via playlist feature',
       7: 'clips uploaded from the local',
       8: 'vertical clips from Youtube Highlight'}
user['clip_type_id']= user['clip_type_id'].map(replacement)
```
