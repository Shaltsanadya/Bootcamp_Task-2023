```python
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

```python
import numpy as np
plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))
heatmap = sns.heatmap(df.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```

```python
plt.figure(figsize=(5, 8))
heatmap = sns.heatmap(df.corr()[['SO2']].sort_values(by='SO2', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Features Correlating with SO2', fontdict={'fontsize':18}, pad=16);
```

insight: 
    1.number of manufacturing enterprises employing 20 or more workers or manu has most correlation with SO2. so the increassing of manu will increase rate of SO2 because industry activity was increased!
    2. temp is strong reason the increassing of SO2 too
    3. The smallest reason or correlation with the increassing of SO2 is precip or average annual precipitation in inches.

```python
sns.set(style='dark')
ax= sns.barplot("SO2", "City", data= df)
fig = plt.gcf()
fig.set_size_inches(30,40)
```

```python
fig, ax = plt.subplots(figsize= (7,6),frameon= True)
ax.plot(df['City'], df['SO2'],c= '#A75D5D', linewidth= 2, marker= '.', markerfacecolor= '#3C6255', markersize= 10)
ax.set_title('Polution')
ax.set_xlabel('SO2')
ax.set_ylabel('City')
ax.tick_params(axis= 'x',rotation=90)
ax.grid(axis= 'y')
ax.set_facecolor('#EAE0DA')
ax.legend(['Total SO2'])
```

```python
df[['SO2','City']].groupby(["City"]).median().sort_values(by='SO2',ascending=False).head(10).plot.bar(color='r')
plt.show()
```

```python
df[['SO2','City']].groupby(["City"]).median().sort_values(by='SO2',ascending=False).tail(10).plot.bar(color='r')
plt.show()
```

Insight: in this figure explained that 'Chicago' and 'providence' have rate of SO2 more than others.
